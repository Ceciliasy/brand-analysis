```python
def crawl_year(year,yid):
    start_idx = 0
    steps = 100
    newresult = []
    while steps:
        url ='http://fortune.com/api/v2/list/'+str(yid)+'/expand/item/ranking/asc/'+str(start_idx)+'/'+str(steps)
        re=requests.get(url,headers = headers)
        newhtml = json.loads(re.text)
        steps = newhtml['count']
        start_idx = start_idx + steps
        for i in range(steps):
            newresult.append(newhtml['list-items'][i]['title'])
    print(year,len(newresult))
    with open(year, 'w') as f:
        for item in newresult:
            f.write("%s\n" % item)
    return newresult
```

```python
def crawl_old_year(year,yid):
    idx = 1
    newresult = []
    flag = 1
    steps = 1
    while flag:
        url = 'http://fortune.com/data/franchise-list/'+str(yid)+'/'+ str(steps)
        re=requests.get(url,headers = headers)
        newhtml = json.loads(re.text)
        flag = len(newhtml['articles'])
        if(flag):
            for i in range(flag):
                newresult.append(newhtml['articles'][i]['title'])
        steps += 1
    print(year,len(newresult))
    with open(year, 'w') as f:
        for item in newresult:
            f.write("%s\n" % item)
    return newresult
```

# clear up to a single file

```python
import os
import pandas as pd
data_dir = '../data/Fortune_500/'
df_list = []
for file in os.listdir(data_dir):
    df = pd.read_csv(data_dir+file,sep='\n',header = None,names=["brand"])
    df['rank'] = df.index+1
    df['year'] = file
    df_list.append(df)
```

```python
single_df = pd.concat(df_list)
```

```python
single_df.to_excel("total_year_fortune.xlsx")
```
