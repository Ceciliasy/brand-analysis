## error storage

```python
contents = []
error= []
error_file ="../data/twitter_s3/2009/tweets.2009-06-02"
with open(error_file,'r') as f:
    for line in f:
        #print(line)
        try:
            line1 = json.loads(line)
        except:
            #import ipdb;ipdb.set_trace()
            error.append(line)
            continue
        if('text' in line1.keys()):
            #print(line)
            contents.append(line1['text'])
```
