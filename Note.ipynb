Corpus Format for GloVe

1. prepare corpus as a single text file with all words separated by one or more spaces or tabs. 
2. If your corpus has multiple documents, the documents (only) should be separated by new line characters.
3. Cooccurrence contexts for words do not extend past newline characters. 
4. Once you create your corpus, you can train GloVe vectors using the following 4 tools. An example is included in demo.sh, which you can modify as necessary.
